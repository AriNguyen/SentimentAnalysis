## 3. Data Preprocessing

```python
df['Business']
```

```python
x = df.to_string(header=False,
                  index=False,
                  index_names=False).split('\n')
vals = [','.join(ele.split()) for ele in x]
print(vals)
```

## 3. Predict

```python
import pandas as pd

text = pd.read_excel('Flowers10k.xlsx')
```

```python
text = df.to_string()
text = ' '.join(vals)
text
```

```python
model = 'ProsusAI/finbert'

output_dir = './'
output = "predictions.csv"
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model, num_labels=3, cache_dir=None)
```

```python
import os
import sys

sys.path.append('./finBERT/finbert')
```

```python
import nltk
nltk.download('punkt')
```

```python
from finBERT.finbert.finbert import predict

predict(text, model, write_to_csv=True, path=os.path.join(output_dir, output))
```
