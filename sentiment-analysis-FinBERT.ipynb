````python
import pandas as pd

text = pd.read_excel('Flowers10k.xlsx')
````

````python
text_new = text['Flowers'].astype(str)
text_s = ""
for col in text_new:
     text_s += col
text_s
````

````python
model = 'ProsusAI/finbert'
output_dir = './'
output = "predictions.csv"
````

````python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model, num_labels=3, cache_dir=None)
````

````python
finbert.predict(text, model, write_to_csv=True, path=os.path.join(output_dir, output))
````
